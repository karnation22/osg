```python
import pandas as pd, numpy as np, warnings,nltk
warnings.filterwarnings(action='ignore', category=UserWarning, module='gensim')
from gensim.models import LdaModel
from gensim.corpora.dictionary import Dictionary
from gensim.parsing.preprocessing import preprocess_string
from nltk.sentiment.vader import SentimentIntensityAnalyzer as SIA
```

```python
BS_Oil_Non_Return_27June2017=pd.read_csv('BS_Oil_Non_Return_27June2017.csv',header=None,error_bad_lines=False)
BS_Oil_Non_Return_27June2017=BS_Oil_Non_Return_27June2017[[0,1]] #just first two columns
BS_Oil_Return_27June2017=pd.read_csv('BS_Oil_Return_27June2017.csv',header=None,error_bad_lines=False)
BS_Oil_Return_27June2017=BS_Oil_Return_27June2017[[0,1]] #just first two columns
BS_RM_Non_Return_27June2017=pd.read_csv('BS_RM_Non_Return_27June2017.csv',header=None,error_bad_lines=False)
BS_RM_Non_Return_27June2017=BS_RM_Non_Return_27June2017[[0,1]]
Core_Oil_Non_Return_27June2017=pd.read_csv('Core_Oil_Non_Return_27June2017.csv',header=None,error_bad_lines=False)
Core_Oil_Non_Return_27June2017=Core_Oil_Non_Return_27June2017[[0,1]]
Core_Oil_Return_27June2017=pd.read_csv('Core_Oil_Return_27June2017.csv',header=None,error_bad_lines=False)
Core_Oil_Return_27June2017=Core_Oil_Return_27June2017[[0,1]]
Core_RM_Non_Return_27June2017=pd.read_csv('Core_RM_Non_Return_27June2017.csv',header=None,error_bad_lines=False)
Core_RM_Non_Return_27June2017=Core_RM_Non_Return_27June2017[[0,1]]
```

```python
#output new .csv file with avg. sentiment; 
def preprocess_plot(pandasDF):
    sia=SIA()
    pandasDF_new = pandasDF.copy()
    pandasDF_new.columns = ['review','review header']
    preproc_split = (lambda text: sia.polarity_scores(text)['compound']) 
    sentiments = pandasDF['review'].apply(preproc_split).values
    assert(len(sentiments)==len(pandasDF_new))
    pandasDF_new['sentiments'] = pd.Series(sentiments)
    pandasDF_new.loc[pandasDF_new['sentiments'] < -0.5 , 'sentiment label'] = "Very Negative"
    pandasDF_new.loc[(pandasDF_new['sentiments'] >= -0.5) & (pandasDF_new['sentiments'] < 0.00), 'sentiment label'] = "Negative"
    pandasDF_new.loc[(pandasDF_new['sentiments'] == 0.00) , 'sentiment label'] = "Neutral"
    pandasDF_new.loc[(pandasDF_new['sentiments'] > 0.00) & (pandasDF_new['sentiments'] < 0.5) , 'sentiment label'] = "Positive"
    pandasDF_new.loc[(pandasDF_new['sentiments'] >= 0.5) , 'sentiment label'] = "Very Positive"
    display(pandasDF_new.head())
    return pandasDF_new

display(BS_Oil_Non_Return_27June2017.head())
new_BS_Oil_Non_Return_27June2017=preprocess_plot(BS_Oil_Non_Return_27June2017)
display(BS_Oil_Return_27June2017.head())
new_BS_Oil_Return_27June2017=preprocess_plot(BS_Oil_Return_27June2017)
display(BS_RM_Non_Return_27June2017.head())
#new_BS_RM_Non_Return_27June2017=preprocess_plot(BS_RM_Non_Return_27June2017)
#display(Core_Oil_Non_Return_27June2017.head())
#new_Core_Oil_Non_Return_27June2017=preprocess_plot(Core_Oil_Non_Return_27June2017)
#display(Core_Oil_Return_27June2017.head())
#new_Core_Oil_Return_27June2017=preprocess_plot(Core_Oil_Return_27June2017)
#display(Core_RM_Non_Return_27June2017.head())
#new_Core_RM_Non_Return_27June2017=preprocess_plot(Core_RM_Non_Return_27June2017)
```
